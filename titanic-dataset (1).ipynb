****Titanic Data Analysis Project:****

Exploring survival patterns using Python and Pandas.

```python
import pandas as pd

df = pd.read_csv("/kaggle/input/competitions/titanic/train.csv")
df.head()
```

1. Understanding the Dataset:

```python
df.shape
```

```python
df.info()
```

2. Checking Missing Values:

```python
df.isnull().sum()
```

3. Overall Survival Rate:

```python
df['Survived'].sum()
```

```python
df['Survived'].mean()
```

4. Survival by Gender:

```python
df.groupby('Sex')['Survived'].mean()
```

5. Survival by Passenger Class:

```python
df.groupby('Pclass')['Survived'].mean()
```

6. Data Visualization:

```python
import matplotlib.pyplot as plt

df['Survived'].value_counts().plot(kind='bar')
plt.title("Survival Count")
plt.show()
```

 7. Detailed Survival Analysis (Sorted Results):

```python
df.groupby('Pclass')['Survived'].mean().sort_values(ascending=False)
```

8. Survival Rate by Gender:

```python
df.groupby('Sex')['Survived'].mean().sort_values(ascending=False)
```

9. Survival Rate Visualization by Class:

```python
import matplotlib.pyplot as plt

df.groupby('Pclass')['Survived'].mean().plot(kind='bar')
plt.title("Survival Rate by Class")
plt.ylabel("Survival Rate")
plt.show()
```

10. Conclusion:

After analyzing the Titanic dataset, the following insights were found:

- Females had significantly higher survival rates compared to males.  
- Passengers in 1st class had the highest survival rate, while 3rd class had the lowest.  
- Social class and gender were strong factors influencing survival chances.  
- This suggests that priority during rescue was given based on gender and socio-economic status.

Overall, the dataset shows clear patterns where survival was not random but influenced by multiple factors such as class, gender, and social position.
